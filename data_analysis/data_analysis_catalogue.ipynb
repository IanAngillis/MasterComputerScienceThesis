This notebook contains the data analysis for RQ1

```python
#Configuration

#Importing libraries
import json
import functools
import matplotlib.pyplot as plt
import seaborn as sns
import copy
import pandas as pd

# Setting color palette for sns
colors = sns.color_palette('pastel')
```

```python
total_identified_smells = 95
identified_maintenance_smells = 54
identified_bloater_smells = 28
identified_security_smells = 11
identified_layer_optimization_smells = 2

labels = ["MAINTENANCE", "BLOATERS", "SECURITY", "LAYER-OPTIMIZATION"]
detections = [identified_maintenance_smells, identified_bloater_smells, identified_security_smells, identified_layer_optimization_smells]

#create pie chart
plt.pie(detections, labels = labels, colors = colors, autopct='%.0f%%')
plt.savefig('./images/saving-a-seaborn-plot-as-pdf-file.pdf')
plt.show()

```

```python
# Subcategories in maintenance
version_pinning = 12
interaction_prevention = 8
maintenance_other = 34

if(version_pinning + interaction_prevention + maintenance_other) == identified_maintenance_smells:
    maintenance_labels = ["VERSION-PINNING", "INTERACTION-PREVENTION", "OTHER"]
    maintenance_detections = [version_pinning, interaction_prevention, maintenance_other]
    plt.pie(maintenance_detections, labels = maintenance_labels, colors = colors, autopct='%.0f%%')
    plt.savefig('./images/maintenance_pie_chart.pdf')
    plt.show()
    
else:
    print("Maintenance data is not consistent")
```

```python
# Subcategories in bloaters
keeping_cache = 12
temporary_files = 8
recommended_packages = 3
bloaters_other = 5

if(keeping_cache + temporary_files + recommended_packages + bloaters_other) == identified_bloater_smells:
    bloaters_labels = ["KEEPING-CACHE", "TEMPORARY-FILES", "RECOMMENDED", "OTHER"]
    bloaters_detections = [keeping_cache, temporary_files, recommended_packages, bloaters_other]
    
    plt.pie(bloaters_detections, labels = bloaters_labels, colors = colors, autopct='%.0f%%')
    plt.savefig('./images/bloaters_pie_chart.pdf')
    plt.show()
else:
    print("Bloaters data is not consistent")
```
